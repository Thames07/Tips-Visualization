```python
import pandas as pd

# visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns


# print the graphs in the notebook
% matplotlib inline

# set seaborn style to white
sns.set_style("white")
```

```python
omo = pd.read_csv('tips.csv')
```

```python
omo
```

```python
omo.head()
```

```python
omo.tail()
```

```python
del omo['Unnamed: 0']

omo.head()
```

```python
# create histogram
ttbill = sns.distplot(omo.total_bill);

# set lables and titles
ttbill.set(xlabel = 'Value', ylabel = 'Frequency', title = "Total Bill")

# take out the right and upper borders
sns.despine()
```

```python
sns.jointplot(x ="total_bill", y ="tip", data = omo)
```

```python
sns.pairplot(omo)
```

```python
sns.stripplot(x = "day", y = "total_bill", data = omo, jitter = True);
```

```python
sns.stripplot(x = "tip", y = "day", hue = "sex", data = omo, jitter = True);
```

```python
sns.boxplot(x = "day", y = "total_bill", hue = "time", data = omo);
```

```python
# better seaborn style
sns.set(style = "ticks")

# creates FacetGrid
g = sns.FacetGrid(omo, col = "time")
g.map(plt.hist, "tip");
```

```python
g = sns.FacetGrid(omo, col = "sex", hue = "smoker")
g.map(plt.scatter, "total_bill", "tip", alpha =.7)

g.add_legend();
```
